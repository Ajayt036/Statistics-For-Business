## 1. BarPlot vs Histogram:

Barplot is used  to represent Quatative data and Historgram is used to represent qualatative data.

1. Barplot: When barplot is made, think of a quantative data, who unique values are on x axis- and any calucytlative value will be shown with w.r.t height of bar on y - axis. this calulation can be count, sum, average etc. Graphs that we get from excel pivot table is called barplot.


2. Histogram takes all the value of a quantative column and put them on x-axis and then shows the frequency of these valye with help of histogrm on y-axis. Note that these values sometimes are continuous and can not be managed on x-Axis. In that case we make bins of value to manage them on X-axis.
