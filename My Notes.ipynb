### 1. Graphical Summaries from Categorical and quantative data:


Histogram - For frequency distribution of values in quant. Column ( mostly used to show value in bins).
<BR>
Barchart = For distribution of values in categorical Column (for abs. Number of values).
<BR>
Dot Plot - Bar chart of Quant columns (shows distribution of values in quant column.) but thos can be replaced by bar chart with width manipulation.
<BR>
Stem and Leaf - A better version of Hsitogram, It shows actual values of data in the form of histogram. best when data values are close to each other.
<BR>
Pie chart = For relative frequency distribution. ( mostly used for Categorical column)
<BR>
CrossTab = To put one varible(qual.) on x and another(quant. in class intervals mostly) on y axis to see summary.
<BR>
Scatter Plot - Used to see relation between two Quantative variables.
<BR>
Side by Side Bar chart - Compare two categorical variables
<BR>
Stacked Bar char: compare relative frequency of two variables.
